```python
import time
from selenium import webdriver
import chromedriver_binary
import pandas as pd
from IPython.core.display import display

url = "https://www.facebook.com/login/device-based/regular/login/?login_attempt=1&amp%3Blwv=101"
email_address = 'ballroom.love.z@gmail.com'
password = '4869Haya'

plofiles = []
urls = []

driver = webdriver.Chrome()
driver.get(url)
time.sleep(5)
userNameField = driver.find_element_by_xpath("//*[@id='email']")
userNameField.send_keys(email_address)

passwordField = driver.find_element_by_xpath("//*[@id='pass']")
passwordField.send_keys(password)

submitButton = driver.find_element_by_xpath("//*[@id='loginbutton']")
submitButton.click()

for a in driver.find_elements_by_class_name('profileLink'):
    plofiles.append(a.text)
    urls.append(a.get_attribute('href'))
    
df = pd.DataFrame.from_dict({"name": plofiles, "url": urls})
display(df)

time.sleep(5)
driver.quit()
```
