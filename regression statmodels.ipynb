```python
#　単回帰分析
import statsmodels.api as sm
import pandas as pd
 
# データを読み込む
data = pd.read_excel(r"C:\Users\ballr\OneDrive\ドキュメント\yachin.xlsx")
 
#回帰分析に使うデータの指定
x = data.iloc[:,[2]] #説明変数
y = data.iloc[:,1] #目的変数
 
#全要素が1の列を説明変数の先頭に追加（絶対必要！！）
X = sm.add_constant(x)
 
#モデルの設定
model = sm.OLS(y, X)
 
#回帰分析の実行
results = model.fit()
 
#結果の詳細を表示
print(results.summary())
```

```python
#　重回帰分析
import statsmodels.api as sm
import pandas as pd
 
# データを読み込む
data = pd.read_excel(r"C:\Users\ballr\OneDrive\ドキュメント\yachin.xlsx")
 
#回帰分析に使うデータの指定
x = data.iloc[:,[2,3,4]] #説明変数
y = data.iloc[:,1] #目的変数
 
#全要素が1の列を説明変数の先頭に追加（絶対必要！！）
X = sm.add_constant(x)
 
#モデルの設定
model = sm.OLS(y, X)
 
#回帰分析の実行
results = model.fit()
 
#結果の詳細を表示
print(results.summary())
```
