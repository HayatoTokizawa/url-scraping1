```python
import requests
from bs4 import BeautifulSoup
import sys
import os
import json
from time import sleep
import urllib
import pandas as pd

crawle_user_id = "UGMD"

retweet_list = []
favorite_list = []
name_list = []

def parseHtml(html_doc):
    soup = BeautifulSoup(html_doc, "html.parser")
    tweet_blocks = soup.select(".tweet")
    for tweet_block in tweet_blocks:
        name = tweet_block.select(".fullname")[0].text
        retweet = tweet_block.select(".ProfileTweet-action--retweet")[0].text
        favorite = tweet_block.select(".ProfileTweet-action--favorite")[0].text
        retweet_list.append(retweet.replace("\n", "").replace("件のリツイート", ""))
        favorite_list.append(favorite.replace("\n", "").replace(" いいね", ""))
        name_list.append(name)

def getMinPosition(html_doc):
    # HTMLを解析
    soup = BeautifulSoup(html_doc, "html.parser")

    # min-positionを抽出
    min_position = soup.select('div[data-min-position]')[0].attrs['data-min-position']

    # 値を返却
    return min_position


def getNextTweet(user_id, max_position):
    # URLの作成
    base_url = 'https://twitter.com/i/profiles/show/'+ user_id +\
    '/timeline/tweets?include_available_features=1&include_entities=1&max_position='+ max_position +\
    '&reset_error_state=false'
    
    # ベースURLからデータを取得
    url_data = urllib.request.urlopen(base_url)

    # JSON形式でデータを取得
    content = json.loads(url_data.read().decode('utf8'))

    # 次取得のIDを記録しておく
    next_position = content['min_position']

    # HTMLをパースしてファイルに書き出し
    parseHtml(content['items_html'])
    print('{}の出力完了'.format(next_position))

    # 一応スリープをかませる
    sleep(1)
    
    # 再起的に呼び出す
    getNextTweet(user_id, next_position)


def getFirstTweet(user_id):
    # ベースURLからデータを取得
    html_doc = requests.get('https://twitter.com/' + user_id).text
    
    # パースして書き出し
    parseHtml(html_doc)

    # 取得開始位置を取得
    max_position = getMinPosition(html_doc)

    # 次
    try:
        getNextTweet(crawle_user_id, max_position)
    except:
        pass



# プログラム実行
getFirstTweet(crawle_user_id)
df2 = pd.DataFrame.from_dict({"投稿者名": name_list, "リツイート数": retweet_list, "いいね数": favorite_list})
df2
```
