```python
import json
import urllib
import requests
import bs4
from time import sleep

dkst = []
dknd = []
fhiow = []

crawle_user_id = 'UGMD'

def parseHtml(html_data):
    soup = bs4.BeautifulSoup(html_data, 'html.parser') # BeautifulSoupã®åˆæœŸåŒ–
    tags = soup.find_all('div',{'class':'content'}) 
    for tag in tags:
        comment = tag.find('p',{'class':'TweetTextSize TweetTextSize--normal js-tweet-text tweet-text'})
        retweet = tag.find('span',{'class':'ProfileTweet-action--retweet u-hiddenVisually'})
        favorite = tag.find('span',{'class':'ProfileTweet-action--favorite u-hiddenVisually'})
        dkst.append(retweet.text)
        dknd.append(favorite.text)
        fhiow.append(comment.text)

def getMinPosition(html_data):
    """
    # æ¬¡ã®å–å¾—é–‹å§‹ä½ç½®ã‚’æŠ½å‡ºã—ã€è¿”ã—ã¾ã™
    #  @params html_data        string          è§£æã™ã‚‹HTMLãƒ‡ãƒ¼ã‚¿ã‚’æ¸¡ã—ã¾ã™
    #  @return                  string          æ¬¡ã®é–‹å§‹ä½ç½®ï¼ˆmin_positionï¼‰ãŒè¿”ã‚Šã¾ã™
    """
    # HTMLã‚’è§£æ
    soup = bs4.BeautifulSoup(html_data)

    # min-positionã‚’æŠ½å‡º
    min_position = soup.select('div[data-min-position]')[0].attrs['data-min-position']

    # å€¤ã‚’è¿”å´
    return min_position


def getNextTweet(user_id, max_position):
    """
    # ä¸ãˆã‚‰ã‚ŒãŸãƒã‚¸ã‚·ãƒ§ãƒ³ã‹ã‚‰å†å¸°çš„ã«ã‚¯ãƒ­ãƒ¼ãƒ«ã‚’å§‹ã‚ã¾ã™
    #  @params user_id      string          Twitterä¸Šã®åŠè§’è‹±æ•°ã®ãƒ¦ãƒ¼ã‚¶åï¼ˆä¾‹ï¼šnnahitoï¼‰
    #  @params max_position string          æ¬¡ã®Twitterã®ã¤ã¶ã‚„ãID
    """
    # URLã®ä½œæˆ
    base_url = 'https://twitter.com/i/profiles/show/'+ user_id +'/timeline/tweets?include_available_features=1&include_entities=1&max_position='+ max_position +'&reset_error_state=false'
1070920401591529472
https://twitter.com/i/profiles/show/UGMD/timeline/tweets?include_available_features=1&include_entities=1&max_position=1070920401591529472&reset_error_state=false
    # ãƒ™ãƒ¼ã‚¹URLã‹ã‚‰ãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—
    url_data = urllib.request.urlopen(base_url)

    # JSONå½¢å¼ã§ãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—
    content = json.loads(url_data.read().decode('utf8'))

    # æ¬¡å–å¾—ã®IDã‚’è¨˜éŒ²ã—ã¦ãŠã
    next_position = content['min_position']

    # HTMLã‚’ãƒ‘ãƒ¼ã‚¹ã—ã¦ãƒ•ã‚¡ã‚¤ãƒ«ã«æ›¸ãå‡ºã—
    parseHtml(content['items_html'])

    # ä¸€å¿œã‚¹ãƒªãƒ¼ãƒ—ã‚’ã‹ã¾ã›ã‚‹
    sleep(1)

    # å†èµ·çš„ã«å‘¼ã³å‡ºã™
    getNextTweet(user_id, next_position)


def getFirstTweet(user_id):
    """
    # ä¸ãˆã‚‰ã‚ŒãŸUserIDã‹ã‚‰ã€æœ€åˆã®Tweetã‚’å–å¾—ã—ã¾ã™
    #  @params user_id      string          å–å¾—ã—ãŸã„ãƒ¦ãƒ¼ã‚¶ã®åŠè§’è‹±æ•°ã®ãƒ¦ãƒ¼ã‚¶åï¼ˆä¾‹ï¼šnnahitoï¼‰
    """
    # ãƒ™ãƒ¼ã‚¹URLã‹ã‚‰ãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—
    url_data = urllib.request.urlopen('https://twitter.com/' + user_id)

    # HTMLãƒ‡ãƒ¼ã‚¿ã‚’å–å¾—
    html_data = url_data.read()

    # ãƒ‘ãƒ¼ã‚¹ã—ã¦æ›¸ãå‡ºã—
    parseHtml(html_data)

    # å–å¾—é–‹å§‹ä½ç½®ã‚’å–å¾—
    max_position = getMinPosition(html_data)

    # æ¬¡
    getNextTweet(crawle_user_id, max_position)

getFirstTweet(crawle_user_id)
print(dkst)
print(dknd)
print(fhiow)
    
    
    #title = tag.text
    #url = query_string_remove(tag.select("a")[0].get("href").replace("/url?q=",""))
```
